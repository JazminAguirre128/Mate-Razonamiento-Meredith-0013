### Problema 3: El número al que se le resta 7

* Planteamento


Resolución MatemáticaEcuación: $x - 7 = 18$Despeje: El 7 que está restando pasa al otro lado sumando.$$x = 18 + 7$$Solución: $x = 25$

```python
resta = int(input("Ingrese el número que resta: "))
resultado = int(input(f"Ingrese el resultado de x - {resta}: "))

# Para hallar x, sumamos lo que se restó
numero = resultado + resta

print(f" X es igual a: {numero}")
```

### Problema 4: El triple de un número más cuatro
Resolución MatemáticaEcuación: $3x + 4 = 31$Despeje (parte 1): Primero movemos lo que está sumando. El 4 pasa restando.$$3x = 31 - 4$$$$3x = 27$$Despeje (parte 2): El 3 que multiplica a $x$ pasa dividiendo.$$x = \frac{27}{3}$$Solución: $x = 9$

```python
# Problema 4: 3x + 4 = 31
total = int(input("Ingrese el total: "))
termino_independiente = int(input("Ingrese el término independiente: "))
coeficiente = int(input("Ingrese el coeficiente: "))

paso1 = total - termino_independiente

x = paso1 / coeficiente

print(f"X es igual a: {int(x)}")
```

### Problema 5: Dimensiones del terreno (Perímetro)
Resolución MatemáticaEcuación planteada: $4x + 10 = 50$ (Donde $x$ es el ancho).Despeje (parte 1): Restamos los 10 metros excedentes del largo al perímetro total.$$4x = 50 - 10$$$$4x = 40$$Despeje (parte 2): Dividimos el resto entre los 4 lados representados en la ecuación.$$x = \frac{40}{4}$$Solución:Ancho ($x$): 10 metros.Largo ($x + 5$): 15 metros.Comprobación: $10 + 10 + 15 + 15 = 50$.

```python
# Pedir el perímetro al usuario
p = float(input("¿Cuánto mide el perímetro total?: "))

# Cálculo matemático directo
ancho = (p - 10) / 4
largo = ancho + 5

# Mostrar resultados
print(f"Cada ancho mide: {ancho} m")
print(f"Cada largo mide: {largo} m")
```
