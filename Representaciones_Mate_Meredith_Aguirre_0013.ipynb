### Priblema 1
Planteamiento: Identificamos la incógnita. "Un número" será nuestra variable $x$. La frase "la suma de un número más 9" se traduce como $x + 9$. El "igual a 21" completa nuestra ecuación.Ecuación:$$x + 9 = 21$$Despeje: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Encontrar el numero desconocido
# Resultado_Total = 21


# Definimos los valores conocidos
resultado_total = int(input("Ingrese el resultado total: "))
numero_sumado = int(input("Ingrese el número sumado: "))

# Despejamos x (x = 21 - 9)
x = resultado_total - numero_sumado

# Mostramos el resultado
print(f"El número buscado es: {x}")
```

Problema 2
Planteamiento:Edad de Ana: La llamaremos $x$ (es nuestra unidad base).Edad de Pedro: El problema dice que tiene "el doble", por lo tanto es $2x$.La suma de ambos es 36.Ecuación:$$x + 2x = 36$$Simplificación: Sumamos las términos semejantes ($1x + 2x$):$$3x = 36$$Despeje: El 3 que multiplica a la $x$ pasa dividiendo al 36:$$x = \frac{36}{3}$$Solución:Ana ($x$): $12$ años.Pedro ($2x$): $2 \times 12 = 24$ años.Comprobación: $12 + 24 = 36$.

```python
# Problema 2: Pedro (2x) y Ana (x) suman 36 años

total_edades = 36

# La ecuación es x + 2x = 36 -> 3x = 36
# Por lo tanto, x = 36 / 3
edad_ana = total_edades / 3
edad_pedro = edad_ana * 2

# Imprimimos resultados (usando int() para ver números enteros)
print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
